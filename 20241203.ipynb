```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'traffic' DataFrame has 'time_period_ending' column
# Convert 'time_period_ending' to timedelta
traffic['time_period_ending'] = pd.to_timedelta(traffic['time_period_ending'])

# Add a column with the hour based on 'time_period_ending'
traffic['hour'] = traffic['time_period_ending'].dt.components['hours']

# Display the modified DataFrame
#traffic[['time_period_ending', 'hour']]

# Plotting the scatter plot
fig, ax = plt.subplots(figsize=(10, 6))

ax.set_xlim(-1, 24)
plt.suptitle('Avg speed during day', fontsize=16, y=0.95)

# Add subtitle with smaller font size
plt.title('Measurement performed every 15 min', fontsize=10, y=0.95)

sns.lineplot(data=traffic, x='hour', y='avg_mph')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib


sns.set_theme()

# Assuming 'traffic' DataFrame has 'time_interval' and 'avg_mph' columns
fig, ax = plt.subplots(figsize=(10, 6))

ax.set_xlim(-1, 96)

sns.scatterplot(data=traffic, x='time_interval', y='avg_mph')

ax.text(x=0.5, y=1.1, s='Avg speed during the day', fontsize=16, weight='bold', ha='center', va='bottom', transform=ax.transAxes)
ax.text(x=0.5, y=1.05, s='Measurements performed evry 15 min', fontsize=10, alpha=0.8, ha='center', va='bottom', transform=ax.transAxes)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter columns with suffix '_cm'
cm_columns = [col for col in traffic.columns if col.endswith('_cm')]

# Melt the DataFrame to long format for seaborn lineplot
melted_traffic = traffic.melt(id_vars=['time_interval'], value_vars=cm_columns, var_name='Measurement', value_name='Value')

# Create the line plot
plt.figure(figsize=(12, 8))
sns.lineplot(data=melted_traffic, x='time_interval', y='Value', hue='Measurement')

plt.title('Line Plot of Measurements Over Time')
plt.xlabel('Time Interval')
plt.ylabel('Measurement Value (cm)')
plt.legend(title='Measurement')
plt.show()
```
