```python
import pandas as pd

agencies = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-02-18/agencies.csv')
```

```python
agencies.head(10)
```

```python
# ...existing code...

# Show all records where 'agency_type' is missing
missing_agency_type = agencies[agencies['agency_type'].isna()]
missing_agency_type
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count plot for 'agency_type'
plt.figure(figsize=(10, 6))
sns.countplot(data=agencies, x='agency_type', order=agencies['agency_type'].value_counts().index)
plt.title('Variation in Agency Type')
plt.xlabel('Agency Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'state' is the column representing the state in the 'agencies' DataFrame
plt.figure(figsize=(15, 10))
sns.countplot(data=agencies, y='state', order=agencies['state'].value_counts().index)
plt.title('Distribution of Agencies by State')
plt.xlabel('Count')
plt.ylabel('State')
plt.show()
```
